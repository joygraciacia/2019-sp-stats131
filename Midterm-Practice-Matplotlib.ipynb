```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, np.pi * 2, 500)
y = np.sin(x)
```

```python
#lineplot
plt.plot(x,y)
plt.title("sin of x")
plt.xlim(0,2*np.pi)
```

```python
x = np.linspace(0, np.pi * 2, 500)
y = np.sin(x)
z = np.tan(x)
w = np.cos(x)

plt.plot(x, y, 'blue')
#plt.plot(x, z, 'red')
plt.plot(x, w, 'orange')
plt.show()

```

```python
plt.subplot(2,2,1)  # number pf rows, number of columns, which plot you want to draw
plt.plot(x, y)
plt.title("sin(x)")
plt.xlabel('x')

plt.subplot(2,2,2) # subplot 2 will be upper right
plt.plot(x, w)
plt.title("cos(x)")


plt.subplot(2,2,3) # subplot 3 will be lower left
plt.plot(x, z)
plt.title("tan(x)")

```
